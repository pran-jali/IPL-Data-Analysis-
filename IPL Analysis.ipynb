Which season consisted of the highest number of matches ever played?

```python
fig_dims = (20, 4)
fig, ax = plt.subplots(figsize=fig_dims)
sns.countplot(x='Season', ax=ax, data=df)
plt.show()
```

**Which is the most successful IPL team with all the data ata hands?**

```python
df1=df.winner.value_counts()
sns.barplot(y = df1.index, x = df1, orient='h')
```

**Which is the probablity of winning a match if the toss was won?**

```python
probability_of_win = df['toss_winner']==df['winner']
probability_of_win.groupby(probability_of_win).size()
```

```python
sns.countplot(probability_of_win)
```
